```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
df_train = pd.read_csv('train.tsv', sep = '\t', header=None)

df_train.info()
df_train.head()
```

```python
import numpy as np

X_train = df_train.values[:, :100]
y_train = df_train.values[:, 100] 

def predict(X, w):
    return X @ w

def mean_squared_error(y_pred, y_true):
    return np.mean((y_true - y_pred)**2)

def LinearRegression(X_train, y_train, epochs=10000, alpha=100):
    N = X_train.shape[0]
    weights = np.random.randn(100)
    
    for epoch in range(epochs):
        y_pred = predict(X_train, weights)
        error = y_pred - y_train
        
        gradient = 2 * X_train.T @ error / N
        
        weights -= alpha * gradient
        
        if epoch % 10 == 0:
            mse = mean_squared_error(y_pred, y_train)
            print(f"Epoch {epoch}, MSE: {mse:.9f}")
    
    return weights
```

```python
weights = LinearRegression(X_train, y_train)
```

```python
df_test = pd.read_csv('test.tsv', sep = '\t', header=None)

df_test.head()
```

```python
answers = df_test.values @ weights
```

```python
answers = pd.DataFrame(answers)
answers.to_csv('answers.tsv', sep='\t', header=None, index=None)
```
